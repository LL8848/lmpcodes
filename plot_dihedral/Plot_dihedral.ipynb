```python
from math import cos

def dihedral(c,fi):
    
    return c[0] + c[1]*cos(fi) + c[2]*cos(fi)**2 + c[3]*cos(fi)**3
```

```python
import matplotlib.pyplot as plt
from math import pi

def plot_dihedral(c1,c2,name):

    x = list(range(0,361))

    v1 = [dihedral(c1,i*pi/180) for i in x]
    v2 = [dihedral(c2,i*pi/180) for i in x]

    plt.plot(x,v1, label = 'OPLS')
    plt.plot(x,v2, label = 'LOPLS')
    
    plt.legend()
    plt.title(name)
```

```python
with open('opls_vs_lopls.prn') as file:
    data = file.readlines()
    
dihe = dict()

for line in data:
    l = line.split()
    dihe[l[0]] = [float(i) for i in l[1:]]


fig, ax = plt.subplots(4,4,sharex=True,figsize=[20,16])


for axi,key in zip(ax.flat,dihe.keys()):
    c1 = dihe[key][0:4]
    c2 = dihe[key][4:8]
    
    x = list(range(0,361))

    v1 = [dihedral(c1,i*pi/180) for i in x]
    v2 = [dihedral(c2,i*pi/180) for i in x]

    axi.plot(x,v1, label = 'OPLS')
    axi.plot(x,v2, label = 'LOPLS')
    axi.title.set_text(key)

plt.savefig('subplots.jpg',dpi=300)
#     plt.legend()
#     plt.title(name)
```
